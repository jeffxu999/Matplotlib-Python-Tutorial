```
# Pie Chart - Matplotlib Python Tutorial

########################################################

import matplotlib.pyplot as plt

pies = [1,1,1,1]

plt.pie(pies)

plt.show()

########################################################

pie_color = ["Orange","Green","LightBlue","Purple"]

pie_label = ["Orange","Green","LightBlue","Purple"]

plt.pie(pies,labels=pie_label,colors=pie_color,autopct="%.0f %%",shadow=True)

plt.axis("equal")

plt.legend()

plt.title("Some Pie!!!")

plt.show()
```
