```
# Error Bar - Matplotlib Python Tutorial

#############################################################

import random

x = [i+1 for i in range(20)]

y = [-i-1 for i in range(20)]

x_error = [random.randint(1,10)/5 for i in range(20)]

y_error = [random.randint(1,10)/5 for i in range(20)]

print(x)

print(y)

print(x_error)

print(y_error)

#############################################################

import matplotlib.pyplot as plt

plt.errorbar(x,y,xerr=x_error,yerr=y_error)

plt.show()

#############################################################

plt.errorbar(x,y,xerr=x_error,yerr=y_error,label="Some Line",color="orange",ecolor="green",capsize=5,capthick=2,errorevery=1)

plt.xlabel("Some X Values")

plt.ylabel("Some Y Values")

plt.title("Some Cool Graph!!!")

plt.legend()

plt.show()
```
