```
# Subplots - Matplotlib Python Tutorial

#############################################################

import matplotlib.pyplot as plt

x = [i for i in range(5)]
y = x

plt.subplot(3,1,1)
plt.plot(x,y,color="red")

plt.subplot(3,1,2)
plt.plot(x,y,color="green")

plt.subplot(3,1,3)
plt.plot(x,y,color="blue")

plt.show()

#############################################################

plt.subplot(1,3,1)
plt.plot(x,y,color="red")

plt.subplot(1,3,2)
plt.plot(x,y,color="green")

plt.subplot(1,3,3)
plt.plot(x,y,color="blue")

plt.show()

#############################################################

plt.subplot(2,3,1)
plt.plot(x,y,color="red")

plt.subplot(2,3,2)
plt.plot(x,y,color="green")

plt.subplot(2,3,3)
plt.plot(x,y,color="blue")

plt.subplot(2,3,4)
plt.plot(x,y,color="orange")

plt.subplot(2,3,5)
plt.plot(x,y,color="purple")

plt.subplot(2,3,6)
plt.plot(x,y,color="black")

plt.show()
```
