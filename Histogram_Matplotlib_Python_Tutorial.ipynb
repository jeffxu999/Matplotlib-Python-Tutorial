```
# Histogram - Matplotlib Python Tutorial

########################################################

import matplotlib.pyplot as plt
import numpy as np

normal_dist = []

for count in range(5000):
    normal_dist.append(np.random.normal(50,15))
    
print(normal_dist)

########################################################

bins = [i for i in range(101)]

plt.hist(normal_dist,bins,histtype="bar")

plt.show()

########################################################

bins = [i*5 for i in range(21)]

plt.hist(normal_dist,bins,histtype="bar",color="orange")

plt.xlabel("Value")

plt.ylabel("Frequency")

plt.title("My First Histogram")

plt.show()

```
