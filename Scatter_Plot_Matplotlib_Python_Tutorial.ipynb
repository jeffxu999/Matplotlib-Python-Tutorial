```
# Scatter Plot - Matplotlib Python Tutorial

########################################################

import matplotlib.pyplot as plt
import numpy as np

x = [i for i in range(500)]

y = []

for number in x:
    random_y = np.random.normal(number,200)
    y.append(random_y)
    
print(x)
print(y)

########################################################

plt.scatter(x,y,color="green",marker="*")

plt.xlabel("X Axis!!!")

plt.ylabel("Y Axis!!!")

plt.title("Super Epic Title!!!")

plt.show()

########################################################

slope, intercept = np.polyfit(x,y,1)

line = []

for x_value in x:
    line_value = x_value*slope + intercept
    line.append(line_value)
    
plt.scatter(x,y,color="orange")
plt.plot(x,line,color="black",linewidth=10)

plt.show()

```
