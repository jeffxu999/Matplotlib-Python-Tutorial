```
# Graph, Legends, Title, & Label - Matplotlib Python Tutorial

########################################################

import matplotlib.pyplot as plt
import random

x = [i for i in range(1,101)]

y = []

for number in x:
    y_value = random.randint(-10,10) + number
    y.append(y_value)
    
print(x)
print(y)

plt.plot(x,y)

plt.show()

########################################################

plt.plot(x,y)

plt.title("Some Cool Graph!!!")

plt.xlabel("This is X Axis")
plt.ylabel("This is Y Axis")

plt.show()

########################################################

x1 = [i for i in range(1,101)]

y1 = []

for number in x1:
    y_value = random.randint(-10,10) + number
    y1.append(y_value)
    
x2 = [i for i in range(1,101)]

y2 = []

for number in x2:
    y_value = random.randint(-20,20) + number
    y2.append(0.5*y_value)
    
plt.plot(x1,y1,label="Line 1")
plt.plot(x2,y2,label="Line 2")

plt.show()
```
