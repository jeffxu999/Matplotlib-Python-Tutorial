```
# 3D Graph - Matplotlib Python Tutorial

#############################################################

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(1,1,1,projection="3d")

x = [np.cos(i) for i in range(100)]
y = [i for i in range(100)]
z = [i for i in range(100)]

ax.plot(x,y,z)

ax.set_xlabel("This is X Axis!")
ax.set_ylabel("This is Y Axis!")
ax.set_zlabel("This is Z Axis!")
plt.title("Some Title!!!")

plt.show()

#############################################################

import random

fig = plt.figure()
ax = fig.add_subplot(1,1,1,projection="3d")

x = [random.randint(1,1000) for i in range(1000)]
y = [random.randint(1,1000) for i in range(1000)]
z = [random.randint(1,1000) for i in range(1000)]

ax.scatter(x,y,z,c="red",marker="v")

ax.set_xlabel("This is X Axis!")
ax.set_ylabel("This is Y Axis!")
ax.set_zlabel("This is Z Axis!")
plt.title("Some Title!!!")

plt.show()

#############################################################

fig = plt.figure()
ax = fig.add_subplot(1,1,1,projection="3d")

x = []
y = []
z = []

for count_macro in range(10):
    x_list = []
    y_list = []
    z_list = []
    
    for count_micro in range(10):
        x_list.append(count_macro)
        y_list.append(count_micro)
        z_list.append(count_micro)
    x.append(np.array(x_list))
    y.append(np.array(y_list))
    z.append(np.array(z_list))
    
x = np.array(x)
y = np.array(y)
z = np.array(z)

print(x)
print(y)
print(z)

ax.plot_surface(x,y,z)

ax.set_xlabel("This is X Axis!")
ax.set_ylabel("This is Y Axis!")
ax.set_zlabel("This is Z Axis!")
plt.title("Some Title!!!")

plt.show()

#############################################################

fig = plt.figure()
ax = fig.add_subplot(1,1,1,projection="3d")

ax.plot_wireframe(x,y,z)

ax.set_xlabel("This is X Axis!")
ax.set_ylabel("This is Y Axis!")
ax.set_zlabel("This is Z Axis!")
plt.title("Some Title!!!")

plt.show()
```
