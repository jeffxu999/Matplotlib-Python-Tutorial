```
# Bar Chart - Matplotlib Python Tutorial

########################################################

import matplotlib.pyplot as plt
import random

x = [i for i in range(1,21)]

y = []

for number in x:
    bar_height = random.randint(2,10)
    y.append(bar_height)
    
print(x)
print(y)

plt.bar(x,y)
plt.show()

########################################################

x1 = [i for i in range(1,41,2)]

y1 = []

for number in x1:
    bar_height = random.randint(2,10)
    y1.append(bar_height)
    
x2 = [i for i in range(2,41,2)]

print(x1)
print(x2)

y2 = [random.randint(2,10)+(i*0) for i in range(1,21)]

plt.bar(x1,y1)
plt.bar(x2,y2)

plt.show()

########################################################

plt.bar(x1,y1,color="orange",width=1)
plt.bar(x2,y2,color="black",width=1)

plt.title("Super Bumblebee Graph")

plt.xlabel("Bumblebee Colonies")
plt.ylabel("Bumblebee Count")

plt.show()
```
